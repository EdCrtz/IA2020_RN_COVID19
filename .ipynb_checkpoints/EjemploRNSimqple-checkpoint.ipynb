El escenario es que tenemos una tienda web que vende libros y otras cosas, y queremos saber si un cliente abandonará una cesta de la compra al momento del pago. Es por eso que estamos haciendo una red neuronal para predecirlo. Para simplificar, todos los datos son solo números. Este será nuestro conjunto de datos. En realidad, puede sustituir esto por cualquier cosa, y siempre que los valores sean números, seguirá funcionando. 
El objetivo es la columna "accion_usuario" y tomamos 1 para indicar que la compra fue exitosa y 0 para indicar que el usuario ha abandonado la canasta.

Ahora vamos a continuar con el archivo de código Python. Todo el código en el resto de esta sección debe colocarse en un solo archivo y colocarse en la misma carpeta que data.csv y new_data.csv. 

```python
# La primera parte del código contiene las declaraciones de importación
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
```

Las primeras cuatro líneas son solo importaciones, las siguientes tres son hiperparámetros. 

*TARGET_VARIABLE* le dice a Python cuál es la variable objetivo que deseamos predecir. 

Ahora debemos dividir los datos de prueba y entrenamiento.
Tenemos un hiperparámetro (TRAIN_TEST_SPLIT) que actualmente tiene el valos 0.5 ( datos en el conjunto de entrenamiento, pero puede cambiar). Tenga cuidado ya que tenemos un pequeño conjunto de datos que podría causar algunos problemas si la división es algo así como 0.95. 

La última línea abre el archivo data.csv.


```python
TARGET_VARIABLE ="accion_usuario"
TRAIN_TEST_SPLIT=0.5
HIDDEN_LAYER_SIZE=10
raw_data = pd.read_csv("data.csv")
```

La primera línea aquí define un muestreo aleatorio de los datos que se utilizarán para obtener la división de los datos de prueba y de entrenamiento. 

Las siguientes dos líneas seleccionan los subconjuntos de datos apropiados del marco de datos original de Pandas (un marco de datos es un objeto similar a una tabla, muy similar a una matriz de Numpy, pero Pandas se enfoca en una fácil remodelación y división, mientras que Numpy se enfoca en un cálculo rápido).
El código para la división de datos de entrenamiento-prueba es:

```python
mask = np.random.rand(len(raw_data)) < TRAIN_TEST_SPLIT
tr_dataset = raw_data[mask]
te_dataset = raw_data[mask]
```

Las siguientes líneas etiquetan los datos datos y etiquetas (labels) y luego los convierten en matrices Numpy, necesarias para que Keras funcione correctamente.

```python
tr_data = np.array(raw_data.drop(TARGET_VARIABLE, axis=1))
tr_labels = np.array(raw_data[[TARGET_VARIABLE]])
te_data = np.array(te_dataset.drop(TARGET_VARIABLE, axis=1))
te_labels = np.array(te_dataset[[TARGET_VARIABLE]])
```

Ahora se especifica en Keras el modelo de la red neuronal, su compilación y entrenamiento .
Necesitamos compilar el modelo, ya que queremos que Keras cree matrices apropiadas para las matrices de peso y sesgo.

La siguiente línea inicializa un nuevo modelo secuencial en una variable llamada ffnn

```python
ffnn = Sequential()
```

Keras es un mdelo de capas. Debemos asegurarnos que la pimera capa tenga el numero datos de entradas
La línea especifica tanto la capa de entrada (en el ejemplo tenemos 3 datos),
y el tamaño de la capa oculta que se especifica al comienzo del archivo en la variable HIDDEN_LAYER_SIZE.
 se especifica la función de activación sigmoide.
 Dense significa que las neuronas de las capas estan completamente conectadas.


```python
ffnn.add(Dense(HIDDEN_LAYER_SIZE, input_shape=(3,),activation="sigmoid"))

## Agregamos otra capa 
ffnn.add(Dense(3, activation="sigmoid"))
```

```python
# La Siguente línea tomará el tamaño de la capa oculta de la capa anterior (Keras lo hace automáticamente) y 
# creará una capa de salida con una neurona.

ffnn.add(Dense(1, activation="sigmoid"))
```

```python
# En seguida se compila el modelo, lo que significa que ensamblará todas las demás cosas que Python necesita
# a partir de lo que hemos especificado
# Al compilar el modelo se mandan llamar las librerias en el backend, en nuestro caso Tensorflow
#  loss --> función de perdida o loss function, que es utilizada para evaluar los pesos.
#  optimizer --> define el oprimizador para los pesos
#  matreics--> reposta la precisión de la clasificación en este caso se usa accuracy

ffnn.compile(loss="mean_squared_error", optimizer="sgd", metrics=["accuracy"])

# entrena la red neuronal con los datos--> tr_data, indicado las etiquetas --> tr_labels, durante 150 épocas, 
# tomando dos muestras en lote. 
# verbose = 1  significa que imprimirá la precisión y la pérdida después
# de cada época de entrenamiento.
# batch_size --> número de instances que son evaluadas antes de que los pesos sean actualizados en la red neuronal. 

ffnn.fit(tr_data, tr_labels, epochs=150, batch_size=1, verbose=1)
```

```python
# evalua el modelo con los datos de prueba
metrics = ffnn.evaluate(te_data, te_labels, verbose=1)

# se imprimen metricas del modelo, en este caso se imprime la certeza del modelo 
print("%s: %.2f%%" % (ffnn.metrics_names[1],metrics[1]*100))
```

A continuación, tomamos el archivo new_data.csv que simula nuevos datos en nuestro sitio web e intentamos predecir lo que sucederá usando el modelo creado:


```python
new_data = np.array(pd.read_csv("new_data2.csv"))
results = ffnn.predict(new_data)
print(results)
```
